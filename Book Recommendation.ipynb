```python
ratings = pd.read_csv('Data\BX-Book-Ratings.csv',sep=';',error_bad_lines=False, encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={
    'User-ID' : 'user_id',
    'Book-Rating': 'ratings'
},inplace=True)
```

```python
# Want users who have rated more than 200 books
ratings['user_id'].value_counts()
```

```python
# Total no of users
ratings['user_id'].unique().shape
```

```python
# users who rated more than 200 books
x = ratings['user_id'].value_counts() >200
```

```python
x[x].shape
```

```python
# y contain userId of those user who rated more than 200 books
y = x[x].index
y
```

```python
# Filter out ratings tables based on user ID i.e. users who rated more than 200 books
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings['user_id'].unique().shape
```

```python
# merge ratings and books
ratings_with_books = ratings.merge(books,on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python
# Analysis
# How many times each book got ratings 
num_ratings = ratings_with_books.groupby('title')['ratings'].count().reset_index()
num_ratings.head()
```

```python
# We want to take those books who got more than 50 ratings
num_ratings.rename(columns={'ratings' : 'num_of_ratings'},inplace=True)
```

```python
num_ratings.head()
```

```python
# merge num_rating with ratings_with_books
final_rating = ratings_with_books.merge(num_ratings,on='title')
```

```python
final_rating.head()
```

```python
final_rating.shape
```

```python
#  We want to take those books who got more than 50 ratings
final_rating = final_rating[final_rating['num_of_ratings'] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.sample(10)
```

```python
final_rating.duplicated().sum()
```

```python
# Convert Data Frame into Pivot table
# In column we put all individual user_id
# In row we put all books title
# In value we place rating corresponding to each user
# In collabarative filtering we want to make clusters of users having similar interest(based on ratings)
# Each row represent Vector corresponding to each row in   space. 
# We have to find out distance between this vectors to make clusters of similar interest 
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values= 'ratings')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
# Convert pivot table into sparse matrix
# So it will consider only non-zero elements during computation
# it will store only non-zero elements in the form of triplets(row,col,value)
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
# Model Building
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(metric='cosine',algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
# 888*1 ----> 1*888
book_pivot.iloc[237,:].values.reshape(1,-1).shape
```

```python
# Test Recommendation system 
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    
   
    for i in suggestion:
        recommend = book_pivot.index[i]
    for i in recommend[1:]:
        print(i)
```
